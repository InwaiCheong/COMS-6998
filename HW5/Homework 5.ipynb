1. Explain the difference between episodic and continuous tasks? Given an example of each.

### Solution:
Episodic tasks has an end state whereas continuous tasks don't, so continuous tasks don't have a reward at the end. Super Mario would be an example of an episodic task as the end state would be the end of a level. A personal assistance robot would be an example of a continuous task as it doesn't have an end state.

2. What do the terms exploration and exploitation mean in RL ? Why do the actors employ $\epsilon$-greedy policy
for selecting actions at each step? Should $\epsilon$ remain fixed or follow a schedule during Deep RL training
? How does the value of $\epsilon$ help balance exploration and exploitation during training.

### Solution:
Exploitation means repeating the same actions since (the agent believes) it maximizes rewards while exporation allows an agent to take a different action that can possibly lead to a higher rewards. Actors employ $\epsilon$-greedy policy as it sort of balances exploration and exploitation and gives the agent a chance to explore new actions instead of repeating the same one, so that the entire action space is explored. The value of $\epsilon$ allows you to decide whether or not the agent will take more explorative vs exploitative approaches i.e: a low epsilon value corresponds to a more exploitative approach and a higher epsilon value makes the agent more explorative. $\epsilon$ should follow a decaying schedule for episodic tasks, from having epsilon decay from $1$ to $0.1$ over time so in the beginning the agent can explore the whole state space by moving randomly and then it will settle down to a fixed exploration rate. But for continuous tasks, a schedule might not be desirable since it might eventually lead to the agent being completely exploitative.

4. What is the benefit of having a target Q-network ?

### Solution:
The benefit of having a target Q-network is that it prevents 'catastrophic forgetting' by having a more stable target network. It allows the network more time to consider more recent actions instead of it updating every single time.

5. How does experience replay help in efficient Q-learning ?

### Solution:
Normally, the algorithm can get 'stuck' due to the next data sample being dependent on the current parameters. By using experience replay, there aren't any oscillations or divergence in parameters as the behavior distribution is averaged over many of its previous states.

6. What is prioritized experience replay ?

### Solution:
Prioritized experience replay is an RL algorithm built on top of experience replay buffers that allows for more efficient learning by replaying important transitions more frequently. This is because not all experience data are equal and some experiences contribute more to than agent's learning than others.

7. Compare and contrast GORILA (General Reinforcement Learning Architecture) and Ape-X architecture. Provide three similarities and three differences.

### Solution:
Similarities: Both can run acting & learning with no high level asynchronization, and the acting & learning algorithms are the same. For both, acting & learning are distributed across multiple workers. <br>

Differences: Ape-X uses a prioritized replay buffer. Ape-X, unlike Gorila, uses a shared, centralized replay memory. Gorila uses uniform sampling, whereas Ape-X samples the most useful data more often.
